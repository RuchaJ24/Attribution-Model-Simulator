```python

def pick_channel(channel_probs):
    """Pick one channel based on probability dict."""
    channels = list(channel_probs.keys())
    probs = list(channel_probs.values())
    return np.random.choice(channels, p=probs)


def generate_channel_sequence(n_touchpoints, scenario):
    """Generate channels with position-aware probabilities."""
    channels = []
    
    for i in range(n_touchpoints):
        if n_touchpoints == 1:
            # Single touch: use first-touch probabilities (it IS first and last)
            ch = pick_channel(scenario['first_touch_probs'])
        elif i == 0:
            # First touch
            ch = pick_channel(scenario['first_touch_probs'])
        elif i == n_touchpoints - 1:
            # Last touch
            ch = pick_channel(scenario['last_touch_probs'])
        else:
            # Middle touch
            ch = pick_channel(scenario['middle_touch_probs'])
        
        channels.append(ch)
    
    return channels

def calculate_conversion_probability(channels, scenario):
    """Calculate conversion probability with scenario-specific logic."""
    
    probability = scenario['base_conversion_rate']
    
    # Component 1: Multi-touch boost (more touches = higher conversion)
    unique_channels = len(set(channels))
    multi_touch_boost = 0.015 * (unique_channels - 1)
    
    # Component 2: Channel quality boost (ground truth embedded here)
    weights = [scenario['ground_truth_weights'].get(ch, 0.1) for ch in channels]
    channel_quality_boost = sum(weights) / len(weights)
    
    # Component 3: First-touch boost (awareness channels)
    first_channel = channels[0]
    first_touch_boost = 0.03 if first_channel in scenario['awareness_channels'] else 0.01
    
    # Component 4: Last-touch boost (conversion channels)
    last_channel = channels[-1]
    last_touch_boost = 0.03 if last_channel in scenario['conversion_channels'] else 0.01
    
    probability = (
        scenario['base_conversion_rate'] +
        multi_touch_boost +
        channel_quality_boost +
        first_touch_boost +
        last_touch_boost
    )
    
    return min(probability, 0.30)

def generate_journey(journey_id, base_date, scenario):
    """Generate one complete customer journey with position-aware channels."""
    
    n_touchpoints = np.random.choice([1, 2, 3, 4, 5], p=scenario['journey_length_probs'])
    
    # Position-aware channel selection
    channels = generate_channel_sequence(n_touchpoints, scenario)
    
    # Timestamps within journey window
    days_offset = np.sort(
        np.random.choice(
            range(scenario['journey_window_days']),
            size=n_touchpoints,
            replace=False
        )
    )
    timestamps = [base_date + timedelta(days=int(d)) for d in days_offset]
    
    # Conversion probability
    conversion_prob = calculate_conversion_probability(channels, scenario)
    converted = np.random.rand() < conversion_prob
    
    # Build records
    touchpoints = []
    for seq_num, (channel, timestamp) in enumerate(zip(channels, timestamps), start=1):
        touchpoints.append({
            'journey_id': journey_id,
            'touchpoint_sequence': seq_num,
            'channel': channel,
            'timestamp': timestamp,
            'converted': converted,
            'conversion_value': 100 if converted else 0
        })
    
    return touchpoints
```

```python
def generate_scenario_data(scenario_key, scenario):
    """Generate complete dataset for one scenario."""
    
    print(f"\n{'='*80}")
    print(f"GENERATING: {scenario['name']}")
    print(f"{'='*80}")
    print(f"Description: {scenario['description']}")
    print(f"Journeys: {scenario['n_journeys']:,}")
    print(f"Avg touches: {scenario['avg_touches']:.2f}")
    print(f"Window: {scenario['journey_window_days']} days")
    print(f"Base conversion: {scenario['base_conversion_rate']:.1%}")
    
    all_touchpoints = []
    base_date = datetime(2024, 1, 1)
    
    print(f"\nGenerating {scenario['n_journeys']:,} journeys...")
    
    for journey_id in range(scenario['n_journeys']):
        if (journey_id + 1) % 5000 == 0:
            print(f"  ✓ {journey_id + 1:,} journeys generated")
        
        touchpoints = generate_journey(journey_id, base_date, scenario)
        all_touchpoints.extend(touchpoints)

    df = pd.DataFrame(all_touchpoints)
    df['days_to_conversion'] = df.groupby('journey_id')['timestamp'].transform(
        lambda x: (x.max() - x.min()).days
    )
    df['total_touchpoints_in_journey'] = df.groupby('journey_id')['touchpoint_sequence'].transform('max')
    
    # Save raw journeys

    raw_path = os.path.join(OUTPUT_DIR, f'{scenario_key}_raw_journeys.csv')
    df.to_csv(raw_path, index=False)
    print(f"\n✓ Saved: {scenario_key}_raw_journeys.csv ({len(df):,} rows)")
    
    # Save journey summary
    summary = df.groupby('journey_id').agg({
        'converted': 'first',
        'conversion_value': 'first',
        'total_touchpoints_in_journey': 'first',
        'days_to_conversion': 'first',
    }).reset_index()
    summary_path = os.path.join(OUTPUT_DIR, f'{scenario_key}_journey_summary.csv')
    summary.to_csv(summary_path, index=False)
    print(f"✓ Saved: {scenario_key}_journey_summary.csv ({len(summary):,} rows)")
    
    # Save ground truth
    gt_df = pd.DataFrame([
        {'channel': ch, 'ground_truth_weight': w}
        for ch, w in scenario['ground_truth_weights'].items()
    ])
    gt_path = f'{OUTPUT_DIR}/{scenario_key}_ground_truth.csv'
    gt_df.to_csv(gt_path, index=False)
    print(f"✓ Saved: {scenario_key}_ground_truth.csv (5 rows)")
    
    # Print validation stats
    print(f"\n--- VALIDATION STATS ---")
    n_conv = summary['converted'].sum()
    conv_rate = n_conv / len(summary)
    print(f"Converting journeys: {n_conv:,} ({conv_rate:.1%})")
    print(f"Total conversion value: ${summary['conversion_value'].sum():,.0f}")
    print(f"Avg touches/journey: {df.groupby('journey_id').size().mean():.2f}")
    
    print(f"\nChannel distribution by position:")
    print(f"\n{'Channel':<15} {'First %':<12} {'Middle %':<12} {'Last %':<12} {'Overall %':<12}")
    print("-" * 65)

    # Get position-wise distribution
    df_sorted = df.sort_values(['journey_id', 'touchpoint_sequence'])
    df_sorted['position_type'] = 'middle'

    # Mark first and last touches
    journey_groups = df_sorted.groupby('journey_id')
    first_touches = journey_groups.head(1).index
    last_touches = journey_groups.tail(1).index

    df_sorted.loc[first_touches, 'position_type'] = 'first'
    df_sorted.loc[last_touches, 'position_type'] = 'last'

    # Single-touch journeys: both first AND last (count as first)
    single_touch_mask = df_sorted['total_touchpoints_in_journey'] == 1
    df_sorted.loc[single_touch_mask, 'position_type'] = 'first'

    for ch in CHANNELS:
        first_pct = (df_sorted[(df_sorted['channel'] == ch) & (df_sorted['position_type'] == 'first')].shape[0] / 
                     df_sorted[df_sorted['position_type'] == 'first'].shape[0]) * 100
        middle_pct = (df_sorted[(df_sorted['channel'] == ch) & (df_sorted['position_type'] == 'middle')].shape[0] / 
                      max(df_sorted[df_sorted['position_type'] == 'middle'].shape[0], 1)) * 100
        last_pct = (df_sorted[(df_sorted['channel'] == ch) & (df_sorted['position_type'] == 'last')].shape[0] / 
                    df_sorted[df_sorted['position_type'] == 'last'].shape[0]) * 100
        overall_pct = (df_sorted[df_sorted['channel'] == ch].shape[0] / len(df_sorted)) * 100
        
        print(f"{ch:<15} {first_pct:>9.1f}%   {middle_pct:>9.1f}%   {last_pct:>9.1f}%   {overall_pct:>9.1f}%")
    return df, summary
```

```python
DATA_DIR = r'D:\Projects\Attribution Model Simulator\data'

def first_touch(journey):
    """100% credit to first channel"""
    first = journey.iloc[0]
    return {first['channel']: first['conversion_value']}


def last_touch(journey):
    """100% credit to last channel"""
    last = journey.iloc[-1]
    return {last['channel']: last['conversion_value']}


def time_decay(journey, decay=0.5):
    """Exponential decay toward last touch"""
    n = len(journey)
    value = journey.iloc[0]['conversion_value']
    weights = [decay ** (n - 1 - i) for i in range(n)]
    total = sum(weights)
    
    credits = {}
    for idx, (_, row) in enumerate(journey.iterrows()):
        ch = row['channel']
        credits[ch] = credits.get(ch, 0) + value * (weights[idx] / total)
    return credits


def position_based(journey):
    """40-20-40 split"""
    n = len(journey)
    value = journey.iloc[0]['conversion_value']
    credits = {}
    
    for idx, (_, row) in enumerate(journey.iterrows()):
        ch = row['channel']
        if n == 1:
            credit = value
        elif idx == 0 or idx == n - 1:
            credit = value * 0.40
        else:
            credit = value * (0.20 / (n - 2))
        credits[ch] = credits.get(ch, 0) + credit
    return credits


def shapley_value(journey):
    """Simplified Shapley: position-weighted marginal contribution"""
    n = len(journey)
    value = journey.iloc[0]['conversion_value']
    credits = {}
    
    for idx, (_, row) in enumerate(journey.iterrows()):
        ch = row['channel']
        pos = idx + 1
        weight = pos / n
        credit = value * weight / n
        credits[ch] = credits.get(ch, 0) + credit
    return credits
```

```python
def run_attribution(scenario_key):
    print(f"\n{'='*80}")
    print(f"ATTRIBUTION ANALYSIS: {scenario_key.upper().replace('_', ' ')}")
    print(f"{'='*80}")
    
    # Load data
    df = pd.read_csv(os.path.join(DATA_DIR, f'{scenario_key}_raw_journeys.csv'))
    gt = pd.read_csv(os.path.join(DATA_DIR, f'{scenario_key}_ground_truth.csv'))
    
    # Get converting journeys
    converted = df[df['converted'] == True].copy()
    print(f"\nConverting journeys: {converted['journey_id'].nunique():,}")
    
    # Initialize results
    results = {
        'First-Touch': {},
        'Last-Touch': {},
        'Time-Decay': {},
        'Position-Based': {},
        'Shapley Value': {}
    }
    
    # Run all models on each journey
    print("Running 5 attribution models...")
    for jid in converted['journey_id'].unique():
        journey = converted[converted['journey_id'] == jid].sort_values('touchpoint_sequence')
        
        # First-Touch
        for ch, val in first_touch(journey).items():
            results['First-Touch'][ch] = results['First-Touch'].get(ch, 0) + val
        
        # Last-Touch
        for ch, val in last_touch(journey).items():
            results['Last-Touch'][ch] = results['Last-Touch'].get(ch, 0) + val
        
        # Time-Decay
        for ch, val in time_decay(journey).items():
            results['Time-Decay'][ch] = results['Time-Decay'].get(ch, 0) + val
        
        # Position-Based
        for ch, val in position_based(journey).items():
            results['Position-Based'][ch] = results['Position-Based'].get(ch, 0) + val
        
        # Shapley
        for ch, val in shapley_value(journey).items():
            results['Shapley Value'][ch] = results['Shapley Value'].get(ch, 0) + val
    
    # Convert to percentages
    channels = sorted(['Paid Search', 'Display', 'Email', 'Organic', 'Social'])
    gt_dict = dict(zip(gt['channel'], gt['ground_truth_weight']))
    
    comparison = {'Ground Truth': {ch: gt_dict.get(ch, 0) * 100 for ch in channels}}
    
    for model_name, model_data in results.items():
        total = sum(model_data.values())
        comparison[model_name] = {
            ch: (model_data.get(ch, 0) / total * 100) if total > 0 else 0
            for ch in channels
        }
    
    # Print comparison table
    print(f"\n{'Channel':<15}", end='')
    for col in ['Ground Truth', 'First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']:
        print(f"{col:<15}", end='')
    print()
    print("-" * 105)
    
    for ch in channels:
        print(f"{ch:<15}", end='')
        for col in ['Ground Truth', 'First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']:
            print(f"{comparison[col][ch]:>5.1f}%        ", end='')
        print()
    
    # Calculate MAPE for each model
    print(f"\n{'Model':<20} {'MAPE':<10} {'Rank':<10}")
    print("-" * 40)
    
    mape_scores = {}
    for model_name in ['First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']:
        errors = [abs(comparison[model_name][ch] - comparison['Ground Truth'][ch]) for ch in channels]
        mape = np.mean(errors)
        mape_scores[model_name] = mape
    
    sorted_models = sorted(mape_scores.items(), key=lambda x: x[1])
    
    for rank, (model, mape) in enumerate(sorted_models, 1):
        marker = "🏆 BEST" if rank == 1 else ""
        print(f"{model:<20} {mape:>5.2f}%    #{rank}  {marker}")
    
    print(f"\n✓ Winner: {sorted_models[0][0]} (MAPE: {sorted_models[0][1]:.2f}%)")
    
    return comparison, mape_scores


# =============================================================================
# RUN BOTH SCENARIOS
# =============================================================================

print("\n" + "="*80)
print("ATTRIBUTION SIMULATOR: MODEL COMPARISON")
print("="*80)

b2b_comparison, b2b_mape = run_attribution('b2b_saas')
consumer_comparison, consumer_mape = run_attribution('consumer_tech')

# =============================================================================
# FINAL SUMMARY
# =============================================================================

print("\n" + "="*80)
print("FINAL VERDICT: BEST MODEL PER SCENARIO")
print("="*80)

print(f"\nB2B SaaS Winner: {min(b2b_mape, key=b2b_mape.get)}")
print(f"  MAPE: {min(b2b_mape.values()):.2f}%")

print(f"\nConsumer Tech Winner: {min(consumer_mape, key=consumer_mape.get)}")
print(f"  MAPE: {min(consumer_mape.values()):.2f}%")

print("\n" + "="*80)
print("✓ ATTRIBUTION ANALYSIS COMPLETE")
print("="*80)
```

```python
# Setup
DATA_DIR = r'D:\Projects\Attribution Model Simulator\data'
OUTPUT_DIR = r'D:\Projects\Attribution Model Simulator\visualizations'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Style
sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['font.family'] = 'sans-serif'

# Color palette
COLORS = {
    'Ground Truth': '#2C3E50',     # Dark blue-grey
    'First-Touch': '#E74C3C',      # Red
    'Last-Touch': '#F39C12',       # Orange
    'Time-Decay': '#3498DB',       # Blue
    'Position-Based': '#9B59B6',   # Purple
    'Shapley Value': '#27AE60'     # Green
}

CHANNELS = ['Display', 'Email', 'Organic', 'Paid Search', 'Social']

# =============================================================================
# RECALCULATE RESULTS (since we need fresh data)
# =============================================================================

def first_touch(journey):
    first = journey.iloc[0]
    return {first['channel']: first['conversion_value']}

def last_touch(journey):
    last = journey.iloc[-1]
    return {last['channel']: last['conversion_value']}

def time_decay(journey, decay=0.5):
    n = len(journey)
    value = journey.iloc[0]['conversion_value']
    weights = [decay ** (n - 1 - i) for i in range(n)]
    total = sum(weights)
    credits = {}
    for idx, (_, row) in enumerate(journey.iterrows()):
        ch = row['channel']
        credits[ch] = credits.get(ch, 0) + value * (weights[idx] / total)
    return credits

def position_based(journey):
    n = len(journey)
    value = journey.iloc[0]['conversion_value']
    credits = {}
    for idx, (_, row) in enumerate(journey.iterrows()):
        ch = row['channel']
        if n == 1:
            credit = value
        elif idx == 0 or idx == n - 1:
            credit = value * 0.40
        else:
            credit = value * (0.20 / (n - 2))
        credits[ch] = credits.get(ch, 0) + credit
    return credits

def shapley_value(journey):
    n = len(journey)
    value = journey.iloc[0]['conversion_value']
    credits = {}
    for idx, (_, row) in enumerate(journey.iterrows()):
        ch = row['channel']
        pos = idx + 1
        weight = pos / n
        credit = value * weight / n
        credits[ch] = credits.get(ch, 0) + credit
    return credits


def calculate_all_models(scenario_key):
    """Run all 5 models on a scenario and return results as DataFrame."""
    df = pd.read_csv(os.path.join(DATA_DIR, f'{scenario_key}_raw_journeys.csv'))
    gt = pd.read_csv(os.path.join(DATA_DIR, f'{scenario_key}_ground_truth.csv'))
    
    converted = df[df['converted'] == True].copy()
    
    results = {
        'First-Touch': {}, 'Last-Touch': {}, 'Time-Decay': {},
        'Position-Based': {}, 'Shapley Value': {}
    }
    
    for jid in converted['journey_id'].unique():
        journey = converted[converted['journey_id'] == jid].sort_values('touchpoint_sequence')
        
        for ch, val in first_touch(journey).items():
            results['First-Touch'][ch] = results['First-Touch'].get(ch, 0) + val
        for ch, val in last_touch(journey).items():
            results['Last-Touch'][ch] = results['Last-Touch'].get(ch, 0) + val
        for ch, val in time_decay(journey).items():
            results['Time-Decay'][ch] = results['Time-Decay'].get(ch, 0) + val
        for ch, val in position_based(journey).items():
            results['Position-Based'][ch] = results['Position-Based'].get(ch, 0) + val
        for ch, val in shapley_value(journey).items():
            results['Shapley Value'][ch] = results['Shapley Value'].get(ch, 0) + val
    
    # Convert to percentages
    gt_dict = dict(zip(gt['channel'], gt['ground_truth_weight']))
    comparison = {'Ground Truth': {ch: gt_dict.get(ch, 0) * 100 for ch in CHANNELS}}
    
    for model_name, model_data in results.items():
        total = sum(model_data.values())
        comparison[model_name] = {
            ch: (model_data.get(ch, 0) / total * 100) if total > 0 else 0
            for ch in CHANNELS
        }
    
    # Calculate MAPE
    mape_scores = {}
    for model_name in ['First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']:
        errors = [abs(comparison[model_name][ch] - comparison['Ground Truth'][ch]) for ch in CHANNELS]
        mape_scores[model_name] = np.mean(errors)
    
    return comparison, mape_scores


print("Loading data and recalculating model results...")
b2b_comparison, b2b_mape = calculate_all_models('b2b_saas')
consumer_comparison, consumer_mape = calculate_all_models('consumer_tech')
print("✓ Done")


# =============================================================================
# VIZ 1: Side-by-Side Scenario Comparison
# =============================================================================
print("\nCreating Viz 1: Side-by-Side Scenario Comparison...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

models_order = ['Ground Truth', 'First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']

# B2B SaaS
x = np.arange(len(CHANNELS))
width = 0.13

for i, model in enumerate(models_order):
    values = [b2b_comparison[model][ch] for ch in CHANNELS]
    ax1.bar(x + i*width, values, width, label=model, color=COLORS[model], edgecolor='white', linewidth=0.5)

ax1.set_title('B2B SaaS: Attribution Model Comparison', fontsize=14, fontweight='bold', pad=15)
ax1.set_ylabel('Attribution %', fontsize=11, fontweight='bold')
ax1.set_xticks(x + width * 2.5)
ax1.set_xticklabels(CHANNELS, fontsize=10)
ax1.legend(fontsize=9, loc='upper right')
ax1.set_ylim(0, 60)
ax1.grid(axis='y', alpha=0.3)

# Consumer Tech
for i, model in enumerate(models_order):
    values = [consumer_comparison[model][ch] for ch in CHANNELS]
    ax2.bar(x + i*width, values, width, label=model, color=COLORS[model], edgecolor='white', linewidth=0.5)

ax2.set_title('Consumer Tech: Attribution Model Comparison', fontsize=14, fontweight='bold', pad=15)
ax2.set_ylabel('Attribution %', fontsize=11, fontweight='bold')
ax2.set_xticks(x + width * 2.5)
ax2.set_xticklabels(CHANNELS, fontsize=10)
ax2.legend(fontsize=9, loc='upper right')
ax2.set_ylim(0, 60)
ax2.grid(axis='y', alpha=0.3)

plt.suptitle('Attribution Model Behavior Across Business Scenarios', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, '1_scenario_comparison.png'), bbox_inches='tight')
plt.show()
print("✓ Saved: 1_scenario_comparison.png")


# =============================================================================
# VIZ 2: MAPE Heatmap (Model × Scenario)
# =============================================================================
print("\nCreating Viz 2: MAPE Heatmap...")

mape_matrix = pd.DataFrame({
    'B2B SaaS': [b2b_mape[m] for m in ['First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']],
    'Consumer Tech': [consumer_mape[m] for m in ['First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']]
}, index=['First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value'])

fig, ax = plt.subplots(figsize=(10, 7))

sns.heatmap(mape_matrix, annot=True, fmt='.2f', cmap='RdYlGn_r', 
            cbar_kws={'label': 'MAPE % (lower = better)'},
            linewidths=2, linecolor='white',
            annot_kws={'fontsize': 14, 'fontweight': 'bold'},
            ax=ax)

# Add winner markers
for col_idx, scenario in enumerate(['B2B SaaS', 'Consumer Tech']):
    winner_idx = mape_matrix[scenario].idxmin()
    winner_row = list(mape_matrix.index).index(winner_idx)
    ax.add_patch(plt.Rectangle((col_idx, winner_row), 1, 1, fill=False, edgecolor='gold', lw=4))

ax.set_title('Model Accuracy by Scenario (MAPE %)\n🏆 Gold border = Winner', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('')
ax.set_ylabel('')
plt.xticks(rotation=0, fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, '2_mape_heatmap.png'), bbox_inches='tight')
plt.show()
print("✓ Saved: 2_mape_heatmap.png")


# =============================================================================
# VIZ 3: The Email Drama (B2B SaaS)
# =============================================================================
print("\nCreating Viz 3: The Email Drama...")

fig, ax = plt.subplots(figsize=(12, 7))

models_list = ['First-Touch', 'Last-Touch', 'Time-Decay', 'Position-Based', 'Shapley Value']
email_predictions = [b2b_comparison[m]['Email'] for m in models_list]
ground_truth_email = b2b_comparison['Ground Truth']['Email']

bars = ax.bar(models_list, email_predictions, color=[COLORS[m] for m in models_list], 
              edgecolor='black', linewidth=1.5)

# Ground truth line
ax.axhline(y=ground_truth_email, color='black', linestyle='--', linewidth=2.5, label=f'Ground Truth ({ground_truth_email:.1f}%)')

# Value labels
for bar, val in zip(bars, email_predictions):
    height = bar.get_height()
    diff = val - ground_truth_email
    color = 'darkred' if abs(diff) > 5 else 'darkgreen' if abs(diff) < 3 else 'darkorange'
    ax.text(bar.get_x() + bar.get_width()/2, height + 1,
            f'{val:.1f}%\n({diff:+.1f}%)', ha='center', va='bottom',
            fontsize=11, fontweight='bold', color=color)

ax.set_title('The Email Problem: How Models Disagree in B2B SaaS', 
             fontsize=15, fontweight='bold', pad=15)
ax.set_ylabel('Email Attribution %', fontsize=12, fontweight='bold')
ax.set_ylim(0, 65)
ax.legend(fontsize=11, loc='upper left')
ax.grid(axis='y', alpha=0.3)

# Add annotation
ax.annotate('Last-Touch says Email drives the business\n(would double-down budget — wrong)',
            xy=(1, 53.9), xytext=(2.5, 60),
            fontsize=10, ha='center',
            bbox=dict(boxstyle='round', facecolor='lightyellow', edgecolor='orange'),
            arrowprops=dict(arrowstyle='->', color='orange'))

ax.annotate('First-Touch says Email is dead\n(would cut budget — also wrong)',
            xy=(0, 2.3), xytext=(0.5, 25),
            fontsize=10, ha='center',
            bbox=dict(boxstyle='round', facecolor='lightyellow', edgecolor='red'),
            arrowprops=dict(arrowstyle='->', color='red'))

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, '3_email_drama.png'), bbox_inches='tight')
plt.show()
print("✓ Saved: 3_email_drama.png")


# =============================================================================
# VIZ 4: Winner Comparison
# =============================================================================
print("\nCreating Viz 4: Winner Comparison...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# B2B SaaS Rankings
b2b_sorted = sorted(b2b_mape.items(), key=lambda x: x[1])
b2b_models = [m[0] for m in b2b_sorted]
b2b_scores = [m[1] for m in b2b_sorted]
colors_b2b = ['#FFD700' if i == 0 else COLORS[m] for i, m in enumerate(b2b_models)]

bars1 = ax1.barh(b2b_models, b2b_scores, color=colors_b2b, edgecolor='black', linewidth=1.5)
for bar, val in zip(bars1, b2b_scores):
    ax1.text(val + 0.2, bar.get_y() + bar.get_height()/2,
             f'{val:.2f}%', va='center', fontsize=11, fontweight='bold')

ax1.text(b2b_scores[0] + 0.5, 0, '🏆 WINNER', va='center', fontsize=13, fontweight='bold', color='darkgoldenrod')
ax1.set_title('B2B SaaS: Model Rankings (MAPE %)', fontsize=14, fontweight='bold')
ax1.set_xlabel('MAPE % (lower = better)', fontsize=11)
ax1.invert_yaxis()
ax1.grid(axis='x', alpha=0.3)
ax1.set_xlim(0, max(b2b_scores) * 1.2)

# Consumer Tech Rankings
consumer_sorted = sorted(consumer_mape.items(), key=lambda x: x[1])
consumer_models = [m[0] for m in consumer_sorted]
consumer_scores = [m[1] for m in consumer_sorted]
colors_consumer = ['#FFD700' if i == 0 else COLORS[m] for i, m in enumerate(consumer_models)]

bars2 = ax2.barh(consumer_models, consumer_scores, color=colors_consumer, edgecolor='black', linewidth=1.5)
for bar, val in zip(bars2, consumer_scores):
    ax2.text(val + 0.2, bar.get_y() + bar.get_height()/2,
             f'{val:.2f}%', va='center', fontsize=11, fontweight='bold')

ax2.text(consumer_scores[0] + 0.5, 0, '🏆 WINNER', va='center', fontsize=13, fontweight='bold', color='darkgoldenrod')
ax2.set_title('Consumer Tech: Model Rankings (MAPE %)', fontsize=14, fontweight='bold')
ax2.set_xlabel('MAPE % (lower = better)', fontsize=11)
ax2.invert_yaxis()
ax2.grid(axis='x', alpha=0.3)
ax2.set_xlim(0, max(consumer_scores) * 1.2)

plt.suptitle('Different Scenarios, Different Winners', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, '4_winner_comparison.png'), bbox_inches='tight')
plt.show()
print("✓ Saved: 4_winner_comparison.png")


# =============================================================================
# VIZ 5: Channel Position Distribution
# =============================================================================
print("\nCreating Viz 5: Channel Position Distribution...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

for ax, scenario_key, title in [(ax1, 'b2b_saas', 'B2B SaaS'), (ax2, 'consumer_tech', 'Consumer Tech')]:
    df = pd.read_csv(os.path.join(DATA_DIR, f'{scenario_key}_raw_journeys.csv'))
    
    # Tag positions
    df_sorted = df.sort_values(['journey_id', 'touchpoint_sequence']).copy()
    df_sorted['position'] = 'middle'
    first_idx = df_sorted.groupby('journey_id').head(1).index
    last_idx = df_sorted.groupby('journey_id').tail(1).index
    df_sorted.loc[first_idx, 'position'] = 'first'
    df_sorted.loc[last_idx, 'position'] = 'last'
    # Single-touch = first only
    single = df_sorted['total_touchpoints_in_journey'] == 1
    df_sorted.loc[single, 'position'] = 'first'
    
    # Calculate channel % by position
    position_dist = {}
    for pos in ['first', 'middle', 'last']:
        pos_data = df_sorted[df_sorted['position'] == pos]
        if len(pos_data) > 0:
            position_dist[pos] = (pos_data['channel'].value_counts(normalize=True) * 100).to_dict()
    
    # Plot
    x = np.arange(len(CHANNELS))
    width = 0.28
    
    first_vals = [position_dist.get('first', {}).get(ch, 0) for ch in CHANNELS]
    middle_vals = [position_dist.get('middle', {}).get(ch, 0) for ch in CHANNELS]
    last_vals = [position_dist.get('last', {}).get(ch, 0) for ch in CHANNELS]
    
    ax.bar(x - width, first_vals, width, label='First Touch', color='#3498DB', edgecolor='white')
    ax.bar(x, middle_vals, width, label='Middle Touches', color='#95A5A6', edgecolor='white')
    ax.bar(x + width, last_vals, width, label='Last Touch', color='#E74C3C', edgecolor='white')
    
    ax.set_title(f'{title}: Channel Position in Journey', fontsize=13, fontweight='bold')
    ax.set_ylabel('% of Touchpoints', fontsize=11, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(CHANNELS, fontsize=10)
    ax.legend(fontsize=10)
    ax.grid(axis='y', alpha=0.3)
    ax.set_ylim(0, 60)

plt.suptitle('Where Each Channel Appears in the Journey', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, '5_channel_positions.png'), bbox_inches='tight')
plt.show()
print("✓ Saved: 5_channel_positions.png")


# =============================================================================
# SUMMARY
# =============================================================================
print("\n" + "="*80)
print("✓ ALL VISUALIZATIONS COMPLETE")
print("="*80)
print(f"\nLocation: {OUTPUT_DIR}")
print("\nFiles created:")
print("  1. 1_scenario_comparison.png - Side-by-side B2B vs Consumer")
print("  2. 2_mape_heatmap.png - Model accuracy heatmap")
print("  3. 3_email_drama.png - Email predictions across models")
print("  4. 4_winner_comparison.png - Model rankings")
print("  5. 5_channel_positions.png - Channel position distribution")
print("="*80)
```
